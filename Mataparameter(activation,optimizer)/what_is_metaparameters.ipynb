<h2>üîç What Are <i>Metaparameters</i>?</h2>

<p><b>Metaparameters</b> (also known as <b>hyperparameters</b>) are settings that control the training process of a machine learning model but are <u>not learned from the data</u>. They are set <b>before</b> training starts.</p>

<hr>

<h3>üìò Definition:</h3>
<blockquote>
<b>Metaparameters</b> are configuration variables external to the model and whose values cannot be estimated from the data. They are usually set manually or optimized through tuning techniques.
</blockquote>

<hr>
