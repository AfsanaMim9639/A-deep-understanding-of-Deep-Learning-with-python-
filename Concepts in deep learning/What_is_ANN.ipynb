<h2>কৃত্রিম নিউরাল নেটওয়ার্ক (ANN)</h2>

<p>
কৃত্রিম নিউরাল নেটওয়ার্ক বা ANN হল একটি <strong>মেশিন লার্নিং মডেল</strong>, যা মানুষের মস্তিষ্কের নিউরনের কাজ অনুকরণ করে তৈরি করা হয়েছে।
এটি বিভিন্ন ধরণের তথ্য বিশ্লেষণ ও পূর্বাভাসের জন্য ব্যবহৃত হয়।
</p>

<h3>ANN-এর গঠন</h3>
<ol>
  <li><strong>ইনপুট লেয়ার (Input Layer):</strong> যেখান থেকে ডেটা মডেলে প্রবেশ করে।</li>
  <li><strong>হিডেন লেয়ার (Hidden Layers):</strong> এক বা একাধিক স্তর, যেখানে তথ্য প্রসেস বা বিশ্লেষণ করা হয়।</li>
  <li><strong>আউটপুট লেয়ার (Output Layer):</strong> যেখানে মডেল চূড়ান্ত সিদ্ধান্ত বা ফলাফল প্রদান করে।</li>
</ol>

<h3>ANN কিভাবে কাজ করে?</h3>
<ul>
  <li>প্রথমে ইনপুট লেয়ার তথ্য গ্রহণ করে।</li>
  <li>তারপর হিডেন লেয়ারগুলিতে ওজন (weights) এবং বায়াস (bias) ব্যবহার করে তথ্য প্রক্রিয়া করা হয়।</li>
  <li>প্রতিটি নিউরনে একটি অ্যাক্টিভেশন ফাংশন (activation function) থাকে, যা সিদ্ধান্ত নিতে সাহায্য করে।</li>
  <li>সবশেষে আউটপুট লেয়ার চূড়ান্ত সিদ্ধান্ত প্রদান করে।</li>
</ul>

<h3>উদাহরণ:</h3>
<p>আপনি যদি একটি ইমেইল স্প্যাম কিনা সেটা নির্ধারণ করতে চান, তাহলে:</p>
<ul>
  <li>ইনপুট হবে: ইমেইলের শব্দ ও গঠন</li>
  <li>প্রসেসিং হবে: হিডেন লেয়ারে তথ্য বিশ্লেষণের মাধ্যমে</li>
  <li>আউটপুট হবে: "স্প্যাম" অথবা "নন-স্প্যাম"</li>
</ul>

<h3>ব্যবহার ক্ষেত্র:</h3>
<ul>
  <li>চেহারা শনাক্তকরণ (Face recognition)</li>
  <li>স্প্যাম ফিল্টারিং</li>
  <li>ভয়েস রিকগনিশন</li>
  <li>ডায়াগনোসিস বা চিকিৎসা নির্ণয়</li>
</ul>


<h2>ANN উদাহরণ: ইমেইল স্প্যাম ডিটেকশন</h2>

<h3>সমস্যা:</h3>
<p>
একটি ইমেইল স্প্যাম কিনা তা শনাক্ত করা।
</p>

<h3>ইনপুট ডেটা:</h3>
<ul>
  <li>ইমেইলের শব্দ সংখ্যা</li>
  <li>ইমেইলে “free”, “buy”, “discount” শব্দ আছে কিনা</li>
  <li>সাবজেক্ট লাইনে ক্যাপিটাল লেটার বেশি আছে কিনা</li>
</ul>

<h3>মডেল গঠন:</h3>
<ol>
  <li><strong>Input Layer:</strong> 3টি ফিচার (উপরের ইনপুট অনুযায়ী)</li>
  <li><strong>Hidden Layer:</strong> 1টি লুকানো স্তর, 4টি নিউরনসহ</li>
  <li><strong>Output Layer:</strong> 1টি নিউরন (আউটপুট: স্প্যাম = 1, না হলে = 0)</li>
</ol>

<h3>প্রসেসিং ধাপ:</h3>
<ol>
  <li>প্রথমে ইনপুট ফিচারগুলো মডেলে প্রবেশ করে</li>
  <li>Hidden layer-এ ওজন (weights) এবং বায়াস (bias) ব্যবহার করে তথ্য বিশ্লেষণ হয়</li>
  <li>Activation function (যেমন: ReLU বা Sigmoid) ব্যবহার করে সিদ্ধান্ত গ্রহণ হয়</li>
  <li>Output layer "স্প্যাম" বা "নন-স্প্যাম" সিদ্ধান্ত দেয়</li>
</ol>

<h3>আউটপুট:</h3>
<ul>
  <li>ইমেইলটি স্প্যাম হলে আউটপুট হবে: <strong>1</strong></li>
  <li>ইমেইলটি স্প্যাম না হলে আউটপুট হবে: <strong>0</strong></li>
</ul>




<h2>কৃত্রিম নিউরাল নেটওয়ার্ক (ANN) এর নিয়মাবলি</h2>

<h3>১. ডেটা ইনপুট:</h3>
<p>প্রথমে ইনপুট লেয়ারে ডেটা প্রদান করা হয়। প্রতিটি ইনপুট একটি নির্দিষ্ট ওজন (weight) সহ একটি নিউরনে পাঠানো হয়।</p>

<h3>২. ওজন (Weights) প্রয়োগ:</h3>
<p>প্রতিটি ইনপুট একটি ওজনের মাধ্যমে পরিমাপিত হয়। ওজন নির্দেশ করে ইনপুটটির প্রভাব কতটুকু হবে।</p>

<h3>৩. বায়াস (Bias) যোগ:</h3>
<p>প্রতিটি নিউরনে একটি বায়াস মান যুক্ত হয়, যা সিদ্ধান্ত প্রক্রিয়ায় নমনীয়তা প্রদান করে।</p>

<h3>৪. নিউরনের ইনপুট সমষ্টি গণনা:</h3>
<p>প্রতিটি নিউরনে মোট ইনপুট = (ওজনযুক্ত ইনপুট + বায়াস) হিসাবে গণনা করা হয়।</p>

<h3>৫. অ্যাক্টিভেশন ফাংশন প্রয়োগ:</h3>
<p>
এই ইনপুট সমষ্টির উপর একটি অ্যাক্টিভেশন ফাংশন (যেমনঃ Sigmoid, ReLU) প্রয়োগ করা হয়, যা নির্ধারণ করে নিউরনটি সক্রিয় (active) হবে কি না।
</p>

<h3>৬. আউটপুট লেয়ারে ফলাফল প্রদান:</h3>
<p>
হিডেন লেয়ার থেকে প্রক্রিয়াকৃত ডেটা আউটপুট লেয়ারে যায় এবং চূড়ান্ত ফলাফল প্রদান করে।
</p>

<h3>৭. ক্ষতি (Loss) নির্ধারণ:</h3>
<p>
প্রেডিক্টেড আউটপুট ও আসল আউটপুটের মধ্যে পার্থক্য হিসাব করে একটি <strong>Loss Function</strong> এর মাধ্যমে ভুল নির্ধারণ করা হয়।
</p>

<h3>৮. ব্যাকপ্রোপাগেশন (Backpropagation):</h3>
<p>
Loss-এর ভিত্তিতে ওজন ও বায়াস আপডেট করা হয়, যাতে পরবর্তী বার ভুল কম হয়। এই প্রক্রিয়াকেই Backpropagation বলা হয়।
</p>

<h3>৯. অপ্টিমাইজার ব্যবহার:</h3>
<p>
Gradient Descent বা Adam এর মতো অপ্টিমাইজার ব্যবহার করে ধীরে ধীরে ওজন/বায়াস আপডেট করা হয় এবং মডেল শেখে (training) করে।
</p>

<h3>১০. পুনরাবৃত্তি (Iterations / Epochs):</h3>
<p>
এই প্রক্রিয়া বহুবার পুনরাবৃত্তি হয় যতক্ষণ না মডেল সঠিকভাবে শেখে এবং কম ভুল করে।
</p>
