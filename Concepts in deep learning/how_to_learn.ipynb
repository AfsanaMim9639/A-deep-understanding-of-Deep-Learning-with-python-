<h2>কৃত্রিম নিউরাল নেটওয়ার্ক (ANN) কীভাবে শেখে?</h2>

<p>একটি ANN মডেল শেখে ইনপুট ডেটা থেকে আউটপুট তৈরি করে এবং ভুল (Error/Loss) দেখে নিজেকে আপডেট করে। নিচে ধাপে ধাপে শেখার প্রক্রিয়া দেখানো হলো:</p>

<h3>১. ফরওয়ার্ড প্রোপাগেশন (Forward Propagation)</h3>
<ul>
  <li>ইনপুট ডেটা মডেলের ইনপুট লেয়ারে প্রবেশ করে।</li>
  <li>তথ্য হিডেন লেয়ার ও আউটপুট লেয়ার দিয়ে অতিক্রম করে।</li>
  <li>প্রতিটি লেয়ারে ওজন (weight), বায়াস (bias) এবং অ্যাক্টিভেশন ফাংশন প্রয়োগ করা হয়।</li>
  <li>মডেল একটি আউটপুট প্রদান করে (যা হয়তো ভুলও হতে পারে)।</li>
</ul>

<h3>২. লস ফাংশন (Loss Function)</h3>
<ul>
  <li>মডেল যেই আউটপুট দিয়েছে সেটি যদি আসল লেবেল (সত্য মান) থেকে আলাদা হয়, তাহলে একটি ভুল বা ত্রুটি হয়।</li>
  <li>এই ত্রুটির পরিমাণ বের করতে আমরা <strong>Loss Function</strong> ব্যবহার করি।</li>
  <li>উদাহরণ: MSE (Mean Squared Error), Binary Crossentropy ইত্যাদি।</li>
</ul>

<h3>৩. ব্যাকপ্রোপাগেশন (Backpropagation)</h3>
<ul>
  <li>মডেল শেখার সবচেয়ে গুরুত্বপূর্ণ ধাপ।</li>
  <li>ত্রুটির ভিত্তিতে, প্রতিটি নিউরনের ওজন এবং বায়াস কতটুকু পরিবর্তন করা দরকার তা হিসাব করা হয়।</li>
  <li>Gradient (ডেরিভেটিভ) ব্যবহার করে জানা যায় কোন দিক থেকে ওজন পরিবর্তন করলে ভুল কমবে।</li>
</ul>

<h3>৪. অপ্টিমাইজেশন (Optimization)</h3>
<ul>
  <li>Gradient Descent বা Adam এর মতো Optimizer ব্যবহার করে ওজন ও বায়াস আপডেট করা হয়।</li>
  <li>উদ্দেশ্য: পরবর্তী বার ভুল কম করা।</li>
</ul>

<h3>৫. পুনরাবৃত্তি (Iterations/Epochs)</h3>
<ul>
  <li>এই পুরো প্রক্রিয়াটি বারবার চালানো হয় (প্রতিটি Epoch-এ)।</li>
  <li>প্রতিবার মডেল একটু একটু করে ভাল ফলাফল দিতে শেখে।</li>
</ul>

<h3>৬. শেখার পর (After Training)</h3>
<ul>
  <li>যখন লস খুব কমে যায়, তখন মডেল বলা যায় “শেখে ফেলেছে”।</li>
  <li>এখন নতুন ডেটা দিলেও মডেল সঠিকভাবে প্রেডিকশন করতে পারে।</li>
</ul>

<p><strong>উপসংহার:</strong> <br>
Forward Propagation ব্যবহার করে মডেল প্রেডিকশন তৈরি করে, আর Backpropagation সেই প্রেডিকশনের ভুল বিশ্লেষণ করে ওজন ও বায়াস আপডেট করে, যাতে পরবর্তীতে মডেল আরও ভালোভাবে শিখতে পারে।
</p>


<h2>কৃত্রিম নিউরাল নেটওয়ার্ক (ANN) এর ভিজুয়ালাইজেশন</h2>

<pre>
               [ X1 ]       [ X2 ]       [ X3 ]
                 |            |            |
                 |            |            |
               -----        -----        -----
              |     |      |     |      |     |     ← Input Layer
              |     |      |     |      |     |
               -----        -----        -----
                 |  \        |  \         |  \
                 |   \       |   \        |   \
              ---------   ---------   ---------
             |         | |         | |         |
             |         | |         | |         |  ← Hidden Layer
             |         | |         | |         |
              ---------   ---------   ---------
                   \        |        /
                    \       |       /
                     \      |      /
                    ---------------
                   |               |
                   |   Prediction  |     ← Output Layer (e.g., 0 or 1)
                   |               |
                    ---------------
</pre>

<p><strong>বর্ণনা:</strong></p>
<ul>
  <li><strong>Input Layer:</strong> X1, X2, X3 হল ইনপুট ফিচার</li>
  <li><strong>Hidden Layer:</strong> এক বা একাধিক হিডেন লেয়ার থাকে, যেখানে নিউরন তথ্য প্রক্রিয়া করে</li>
  <li><strong>Output Layer:</strong> মডেলের চূড়ান্ত সিদ্ধান্ত/ফলাফল (যেমনঃ ক্লাস 0 বা 1)</li>
</ul>
