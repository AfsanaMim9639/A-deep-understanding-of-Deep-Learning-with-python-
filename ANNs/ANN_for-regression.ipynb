<h1>Regression-এর জন্য Artificial Neural Network (ANN)</h1>

<hr>

<h2>Regression কী?</h2>
<p>
Regression হলো এমন একটি Machine Learning Task, যেখানে আমরা একটি সংখ্যা বা ধারাবাহিক মান (continuous value) পূর্বানুমান করি। উদাহরণ: বাড়ির দাম, ছাত্রের স্কোর, তাপমাত্রা ইত্যাদি।
</p>

<hr>

<h2>Regression-এর জন্য ANN কেমন হয়?</h2>

<ul>
  <li><b>Input Layer:</b> ইনপুট ফিচার অনুযায়ী নিউরন থাকে।</li>
  <li><b>Hidden Layers:</b> ১ বা একাধিক হিডেন লেয়ার থাকে — Activation Function সহ।</li>
  <li><b>Output Layer:</b> মাত্র ১টি নিউরন থাকে কারণ একটি সংখ্যা প্রেডিক্ট করতে হয়।</li>
  <li><b>Activation Function (Output):</b> সাধারণত <code>Linear (f(x)=x)</code></li>
</ul>

<hr>

<h2>Activation Function নির্বাচন:</h2>

<h2>Loss Function:</h2>
<p>
Regression-এর জন্য সবচেয়ে জনপ্রিয় Loss Function হলো:
</p>

<pre>
Mean Squared Error (MSE) = (1/m) * Σ (y − ŷ)<sup>2</sup>
</pre>

<ul>
  <li>y = আসল মান (Actual value)</li>
  <li>ŷ = পূর্বানুমান (Predicted value)</li>
  <li>m = মোট স্যাম্পলের সংখ্যা</li>
</ul>

<hr>

<h2>উদাহরণ: বাড়ির দাম প্রেডিকশন</h2>

<p>ধরুন আমাদের ফিচার: আয়তন, বেডরুম সংখ্যা, লোকেশন স্কোর</p>

<ul>
  <li><b>ইনপুট:</b> [1200, 3, 7.5]</li>
  <li><b>আউটপুট:</b> 32 লাখ টাকা</li>
</ul>

<p>আমাদের ANN মডেল এটি শিখে ভবিষ্যতের নতুন ইনপুটের জন্য দাম প্রেডিকশন করবে।</p>

<hr>

<h2>সারাংশ:</h2>

<ul>
  <li>Regression-এর জন্য ANN খুব কার্যকর, বিশেষ করে complex nonlinear সম্পর্ক থাকলে।</li>
  <li>Output লেয়ারে <b>Linear activation</b> ব্যবহার করতে হয়।</li>
  <li>Loss Function হিসেবে <b>MSE</b> বা <b>MAE</b> ব্যবহার হয়।</li>
</ul>
