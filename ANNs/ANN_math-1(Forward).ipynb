<h1>Artificial Neural Network (ANN) - Forward Propagation এর গাণিতিক ব্যাখ্যা</h1>

<h2>Forward Propagation কী?</h2>
<p>
Forward Propagation হলো এমন একটি ধাপ, যেখানে ইনপুট ডেটা ধাপে ধাপে নিউরাল নেটওয়ার্কের প্রতিটি লেয়ারের মাধ্যমে পার হয়ে আউটপুট তৈরি করে। প্রতিটি লেয়ারে ইনপুট ডেটার উপর ওজন প্রয়োগ করা হয়, বায়াস যোগ করা হয়, এবং অ্যাক্টিভেশন ফাংশন ব্যবহার করে একটি আউটপুট তৈরি হয়।
</p>

<hr>

<h2>ধাপ ১: Weighted Sum</h2>
<p>
প্রতিটি নিউরনের জন্য প্রথমে ওজনযুক্ত ইনপুট এবং বায়াস যোগ করে একটি সমষ্টি হিসেব করা হয়।
</p>

<p>
গণিতিকভাবে:<br>
<code>z = w<sub>1</sub>x<sub>1</sub> + w<sub>2</sub>x<sub>2</sub> + ... + w<sub>n</sub>x<sub>n</sub> + b</code><br>
বা<br>
<code>z = w · x + b</code>
</p>

<ul>
  <li>x = ইনপুট ফিচার ভেক্টর</li>
  <li>w = ওজন ভেক্টর</li>
  <li>b = বায়াস</li>
</ul>

<hr>

<h2>ধাপ ২: Activation Function</h2>
<p>
Weighted sum-এর উপর একটি অ্যাক্টিভেশন ফাংশন প্রয়োগ করা হয় যা আউটপুট তৈরি করে:
</p>

<p>
<code>a = σ(z)</code><br>
যেখানে <code>σ()</code> হলো অ্যাক্টিভেশন ফাংশন, যেমন:
</p>

<ul>
  <li>Sigmoid: σ(z) = 1 / (1 + e<sup>−z</sup>)</li>
  <li>ReLU: σ(z) = max(0, z)</li>
  <li>tanh: σ(z) = (e<sup>z</sup> − e<sup>−z</sup>) / (e<sup>z</sup> + e<sup>−z</sup>)</li>
</ul>

<hr>

<h2>Hidden Layers এর জন্য সাধারণ সূত্র</h2>
<p>
প্রতিটি হিডেন লেয়ারে নিচের মত প্রক্রিয়া চলে:
</p>

<p>
<code>z<sup>[l]</sup> = W<sup>[l]</sup> · a<sup>[l−1]</sup> + b<sup>[l]</sup></code><br>
<code>a<sup>[l]</sup> = σ(z<sup>[l]</sup>)</code>
</p>

<ul>
  <li>l = লেয়ার নম্বর</li>
  <li>a<sup>[0]</sup> = ইনপুট ডেটা</li>
  <li>a<sup>[l]</sup> = ঐ লেয়ারের আউটপুট (অ্যাক্টিভেশন)</li>
</ul>

<hr>

<h2>Output Layer</h2>
<p>
শেষ লেয়ারে:
</p>

<p>
<code>z<sup>[L]</sup> = W<sup>[L]</sup> · a<sup>[L−1]</sup> + b<sup>[L]</sup></code><br>
<code>ŷ = a<sup>[L]</sup> = σ(z<sup>[L]</sup>)</code>
</p>

<p>এটাই চূড়ান্ত আউটপুট বা প্রেডিকশন।</p>

<ul>

  <li>প্রতিটি ধাপে ওজন, বায়াস ও অ্যাক্টিভেশন ফাংশনের মাধ্যমে ইনফরমেশন প্রক্রিয়াকরণ হয়।</li>
