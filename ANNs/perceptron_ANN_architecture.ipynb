# Perceptron ও Artificial Neural Network (ANN) আর্কিটেকচার

## ১. Perceptron
**Perceptron কী?**  
Perceptron হলো সবচেয়ে সহজ ধরনের কৃত্রিম নিউরাল নেটওয়ার্ক এবং এটি ANN এর বেসিক ইউনিট। এটি একটি সিঙ্গেল-লেয়ার নিউরাল নেটওয়ার্ক।

### আর্কিটেকচার:
- **ইনপুট লেয়ার:** একাধিক ইনপুট নেয়।
- **ওজন (Weights):** প্রতিটি ইনপুটের সাথে একটি ওজন গুণ করা হয়।
- **সমষ্টি (Summation):** ওজনযুক্ত ইনপুটগুলো যোগ করা হয়।
- **বায়াস (Bias):** একটি বায়াস যোগ করা হয়।
- **অ্যাক্টিভেশন ফাংশন:** ফলাফলকে একটি step function এর মাধ্যমে ০ বা ১ এ রূপান্তর করা হয়।

**গণিতিক ফর্মুলা:**  
* y = activation(∑ (w_i * x_i) + b)  
যেখানে,  
x_i = ইনপুট, w_i = ওজন, b = বায়াস।

**ব্যবহার:**  
প্রধানত দ্বিমুখী শ্রেণীবিভাজনের জন্য ব্যবহৃত।

## ২. Artificial Neural Network (ANN) আর্কিটেকচার
**ANN কী?**  
এটি একটি নিউরাল নেটওয়ার্ক যা একাধিক নিউরনের লেয়ার নিয়ে গঠিত। এটি জটিল প্যাটার্ন এবং সম্পর্ক শিখতে পারে।

### লেয়ারসমূহ:
- **ইনপুট লেয়ার:** ইনপুট ফিচার নেয়।
- **হিডেন লেয়ার:** ইনপুট ও আউটপুট লেয়ারের মাঝখানে থাকে, ফিচার বের করে।
- **আউটপুট লেয়ার:** চূড়ান্ত আউটপুট তৈরি করে।

**নিউরনসমূহ:**  
প্রতিটি নিউরন ইনপুটের weighted sum করে, বায়াস যোগ করে, এবং একটি non-linear অ্যাক্টিভেশন ফাংশন প্রয়োগ করে।

**অ্যাক্টিভেশন ফাংশন:**  
যেমন ReLU, sigmoid, tanh ইত্যাদি, যা নেটওয়ার্ককে জটিল সম্পর্ক শিখতে সাহায্য করে।

**ফিডফরোয়ার্ড প্রসেস:**  
ইনপুট ডেটা লেয়ারগুলো হয়ে আগানো হয় এবং আউটপুট তৈরি হয়।

**ব্যাকপ্রোপাগেশন:**  
ভুল কমাতে ওজনগুলো আপডেট করা হয় gradient descent ব্যবহার করে।
